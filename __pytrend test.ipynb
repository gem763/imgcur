```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw = 'photo'
```

```python
kw_list = [kw]
pytrends.build_payload(kw_list, timeframe='now 4-H')#, cat=0, timeframe='today 1-y')
```

```python
qrys = pytrends.related_queries(); qrys;
```

```python
qrys[kw]['top']
```

```python
qrys[kw]['rising']
```

```python
' '.join(qrys[kw]['top']['query'].values).replace('photo', '').replace('  ', ' ').strip()
```

```python
topics = pytrends.related_topics(); topics;
```

```python
topics[kw]['top']
```

```python
topics[kw]['rising']
```

```python
' '.join(topics[kw]['top']['topic_title'].values)
```

```python
' '.join(topics[kw]['rising']['topic_title'].values)
```
